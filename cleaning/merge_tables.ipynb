```python
import pickle
import pandas as pd
```

```python
## load scraped reviews data
reviews = pickle.load(open("../data/hotel_reviews_final.pkl", "rb"))
reviews = reviews.rename(columns={'url': 'link'})
print(reviews.shape)
reviews.head()
```

```python
## check info
reviews.info()
```

```python
## load hotel info
hotels = pickle.load(open("../data/hotel_info_final.pkl", "rb"))
print(hotels.shape)
hotels.head()
```

```python
hotels.info()
```

```python
## merge the two tables
all = hotels.merge(reviews, on = 'link', how = 'outer')
print(all.shape)
all.info()
```

```python
## load hotel location info
location = pickle.load(open("../data/hotel_surroundings_features.pkl", "rb"))
print(location.shape)
location = location.drop(['url', 'location'], axis =1)
location.head()
```

```python
## merge the two tables
all = all.merge(location, on = 'hotel_name_', how = 'outer')
print(all.shape)
all.info()
```

```python
print("Unique hotel names in hotel data: " + str(hotels['hotel_name_'].nunique()))
print("Unique hotel names in combined data: " + str(all['hotel_name_'].nunique()))
print("\n")
print("Unique links in reviews data: " + str(reviews['link'].nunique()))
print("Unique links in hotels data: " + str(hotels['link'].nunique()))
print("Unique links in combined data: " + str(all['link'].nunique()))
print("\n")
print("Shape of reviews data: " + str(reviews.shape))
print("Shape of hotels data: " + str(hotels.shape))
print("Shape of combined data: " + str(all.shape))
print("\n")
print("All Column names: \n" + str(all.columns))
```

```python
## save to pickle file
pickle.dump(all, open('../data/merged_hotel_data.pkl', "wb"))
```
