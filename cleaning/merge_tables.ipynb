```python
import pickle
import pandas as pd
```

```python
## load scraped reviews data
reviews = pickle.load(open("../data/hotel_reviews_final.pkl", "rb"))
reviews = reviews.rename(columns={'url': 'link'})
print(reviews.shape)
reviews.head()
```

```python
## check info
reviews.info()
```

```python
## load hotel info
hotels = pickle.load(open("../data/hotel_info_final.pkl", "rb"))
print(hotels.shape)
hotels.head()
```

```python
hotels.info()
```

```python
## merge the two tables
all = hotels.merge(reviews, on = 'link', how = 'outer')
print(all.shape)
all.info()
```

```python
print("Unique hotel names in hotel data: " + str(hotels['hotel_name_'].nunique()))
print("Unique hotel names in combined data: " + str(all['hotel_name_'].nunique()))
print("\n")
print("Unique links in reviews data: " + str(reviews['link'].nunique()))
print("Unique links in hotels data: " + str(hotels['link'].nunique()))
print("Unique links in combined data: " + str(all['link'].nunique()))
print("\n")
print(reviews.shape)
print(hotels.shape)
print(all.shape)
```

```python
## save to pickle file
pickle.dump(all, open('../data/merged_hotel_data.pkl', "wb"))
```
