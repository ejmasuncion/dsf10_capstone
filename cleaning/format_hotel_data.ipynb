```python
import pickle
import pandas as pd
```

```python
test = pickle.load(open("../scraping/data/dataset_0-425.pkl", "rb"))
test = test.reset_index(drop = True)
print(test.shape)
test
```

```python
temp = test[['hotel_name_', 'room_type', 'price_list']]
temp.head()
```

```python
def get_lowest_prices(x):
    room_type = x['room_type']
    price_list = x['price_list']
    return {room_type[i]: price_list[i] for i in reversed(range(len(room_type)))}
```

```python
temp['room_price_dict'] = temp.apply(get_lowest_prices, axis=1)
```

```python
df = test.merge(temp[['hotel_name_', 'room_price_dict']], on = 'hotel_name_', how = 'left')
df.head()
```
