```python
import pickle
import pandas as pd
import ast
```

```python
## data with urls used in scraping reviews 
urls = pd.read_csv("../scraping/data/consolidated_urls.csv")
```

```python
## loading scraped hotel info data
test = pickle.load(open("../scraping/data/dataset_0-425.pkl", "rb"))
test = test.reset_index(drop = True)

## filter to only include urls in the scraped reviews
## we had instances where there were multiple urls for the same hotel name
test = test[test['link'].isin(urls['link'])]
print(test.shape)
test.head()
```

```python
## creating temporariy dataset to create room-price dict
temp = test[['hotel_name_', 'room_type', 'price_list']]
# temp['room_type'] = temp['room_type'].astype(str)
# temp['price_list'] = temp['price_list'].astype(str)
# temp = temp.drop_duplicates()
print(temp.shape)
temp.head()
```

```python
# temp['room_type'] = temp['room_type'].apply(lambda x: ast.literal_eval(x))
# temp['price_list'] = temp['price_list'].apply(lambda x: ast.literal_eval(x))
```

```python
## function to create a dictionary of room:price
## lowest price is selected if there are multiple prices on a room type
def get_lowest_prices(x):
    room_type = x['room_type']
    price_list = x['price_list']
    return {room_type[i]: price_list[i] for i in reversed(range(len(room_type)))}
```

```python
## apply the function
temp['room_price_dict'] = temp.apply(get_lowest_prices, axis=1)
```

```python
temp.shape
```

```python
print("Unique hotel names in reviews data: " + str(temp['hotel_name_'].nunique()))
print("Unique hotel names in hotels data: " + str(test['hotel_name_'].nunique()))
```

```python
## merge the create the final dataset
df = test.merge(temp[['hotel_name_', 'room_price_dict']], on = ['hotel_name_'], how = 'left')
print(df.shape)
df.head()
```

```python
## save as pkl file
pickle.dump(df, open('../data/hotel_info_final2.pkl', "wb"))
```
