```python
import pandas as pd
import pickle
import seaborn as sns
from matplotlib import pyplot as plt
import itertools
from collections import Counter
import numpy as np
```

```python
## load scraped data
data = pickle.load(open("../data/hotel_info_final2.pkl", "rb"))
data = data.rename(columns={'url': 'link'})
print(data.shape)
data.head()
```

### Exploring facilities

```python
data['all_facilities'][2]
```

```python
# getting all lists in all_facilities
df = []
for i in range(len(data)):
    temp_list = [item for items in data['all_facilities'][i] for item in items]
    df.append(temp_list)
```

```python
# all values in the all_facilities column
df
```

```python
# combine in a single list
all_facilities = [item.lower() for items in df for item in items]
print(len(all_facilities))
all_facilities
```

```python
# number of unique facilities
len(Counter(all_facilities).keys())
```

```python
## languages found in the all_facilities column
languages = ['english', 'filipino', 'japanese', 'chinese', 'korean', 'farsi', 'arabic', 'german', 'spanish', 'french']
```

```python
## removing languages from facilities list
facilities_less_lang =[x for x in all_facilities if x not in languages]
# number of facilities minus languages
len(Counter(facilities_less_lang).keys())
```

```python
facilities_dict = dict(reversed(sorted(Counter(facilities_less_lang).items(), key=lambda item: item[1])))
```

```python
len(facilities_dict)
```

```python
# get most common facilities (top 50)
mini_dict = dict(itertools.islice(facilities_dict.items(),50))
mini_dict
```

```python
# create a list of the top n common facilities sorted by most common
top_facilities = [key for key in mini_dict]
top_facilities
```

```python
mini_df = data[['hotel_name_', 'all_facilities']]
```

```python
mini_df['unlist'] = mini_df['all_facilities'].apply(lambda x: [item.lower() for items in x for item in items])
mini_df['facilities_count'] = mini_df['unlist'].apply(lambda x: len(Counter(x).keys()))
mini_df['in_top_facilities'] = mini_df['unlist'].apply(lambda x: [x for x in x if x in top_facilities])
```

```python
mini_df
```

```python
## creating wide dataset where each new column is a flag if a facility exists in the property
facilities_wide = mini_df.copy()
for col in top_facilities:
    facilities_wide[col] = 0 #temporary value
    for i in range(len(facilities_wide)):
        if len([x for x in facilities_wide['in_top_facilities'][i] if x in col]):
            facilities_wide[col][i] = 1
        else:
            facilities_wide[col][i] = 0
```

```python
facilities_wide
```

```python
#checking counts, should match values in mini_dict
facilities_wide['fire extinguishers'].value_counts()
```

```python
## save to pickle file
pickle.dump(facilities_wide, open('../data/facilities_wide.pkl', "wb"))
```

```python
languages_wide = mini_df.copy()
for col in languages:
    languages_wide[col] = 0 #temp value
    for i in range(len(languages_wide)):
        if len([x for x in languages_wide['unlist'][i] if x in col]):
            languages_wide[col][i] = 1
        else:
            languages_wide[col][i] = 0
```

```python
languages_wide = languages_wide.drop(['in_top_facilities', 'all_facilities', 'unlist'], axis = 1)
```
