```python
import pandas as pd
import pickle
import seaborn as snsa
from matplotlib import pyplot as plt
```

```python
## load scraped data
data = pickle.load(open("../data/hotel_info_final2.pkl", "rb"))
data = data.rename(columns={'url': 'link'})
print(data.shape)
data.head()
```

```python
## load scraped data
data2 = pickle.load(open("../data/hotel_reviews_final.pkl", "rb"))
data2 = data2.rename(columns={'url': 'link'})
print(data2.shape)
data2.head()
```

```python
data2 = data2.applymap(lambda x: x.strip() if isinstance(x, str) else x)
data2.head()
```

### Put two lists into dictionary

```python
room_dict = pd.Series()
data['room_dict']=room_dict
data
```

```python
for i in range(len(data['hotel_name_'])):
    data['room_dict'][i]=[dict(zip(data['room_type'][i], data['price_list'][i]))]
```

```python
df = []
for i in range(len(data['hotel_name_'])):
    hotel_link = data['link'][i]
    hotel_name = data['hotel_name_'][i]
    sub_data = data['room_dict'][i][0]
    temp  = pd.DataFrame.from_dict(sub_data, orient = 'index').reset_index()
    temp['link'] = hotel_link
    temp['hotel_name_'] = hotel_name
    temp.rename(columns = {0:'price', 'index':'room_type'}, inplace = True)
    df.append(temp)
    
```

```python
room_price_data = pd.concat(df)
room_price_data.to_csv("room_price_data.csv")
```

```python
room_price_data[room_price_data['hotel_name_']=="RedDoorz near East Avenue Medical Center"]
```

# Price per Room Type

```python
df2 = []
for i in range(len(data['link'])):
    hotel_link2 = data['link'][i]
    hotel_name2 = data['hotel_name_'][i]
    sub_data2= data['room_price_dict'][i]
    temp2  = pd.DataFrame.from_dict(sub_data2, orient = 'index').reset_index()
    temp2['link'] = hotel_link2
    temp2['hotel_name_'] = hotel_name2
    temp2.rename(columns = {0:'price', 'index':'room_type'}, inplace = True)
    df2.append(temp2)
    
```

```python
room_price_data2 = pd.concat(df2)
# room_price_data2.to_csv("room_price_data2.csv")
```
